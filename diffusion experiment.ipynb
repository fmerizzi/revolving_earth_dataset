```python
import torch
from video_diffusion_pytorch import Unet3D, GaussianDiffusion, Trainer

model = Unet3D(
    dim = 64,
    dim_mults = (1, 2, 4, 8),
)

diffusion = GaussianDiffusion(
    model,
    image_size = 64,
    num_frames = 10,
    timesteps = 10,   # number of steps
    loss_type = 'l1'    # L1 or L2
).cuda()

```

```python
trainer = Trainer(
    diffusion,
    '/home/fmerizzi/Documents/video-diffusion-pytorch/revolving_64',                         # this folder path needs to contain all your training data, as .gif files, of correct image size and number of frames
    train_batch_size = 2,
    train_lr = 1e-4,
    save_and_sample_every = 1000,
    train_num_steps = 2000,         # total training steps
    gradient_accumulate_every = 2,    # gradient accumulation steps
    ema_decay = 0.995,                # exponential moving average decay
    amp = False                        # turn on mixed precision
)

trainer.train()
```

```python
trainer.load("bueno11000")
```

```python
sampled_videos = diffusion.sample(batch_size = 10)
sampled_videos.shape # (4, 3, 5, 32, 32)
```

```python
for i in range(10):

    test = sampled_videos[i].cpu()
    test = test.permute((1,2,3,0))

    import datetime;

    # ct stores current time
    ct = datetime.datetime.now()
    import random
    import numpy as np

    import matplotlib
    import matplotlib.pyplot as plt

    import matplotlib.animation as animation


    fps = 5
    nSeconds = 2
    snapshots = [ test[i] for i in range(len(test)) ]

    # First set up the figure, the axis, and the plot element we want to animate
    fig = plt.figure( figsize=(3,3) )

    a = snapshots[0]
    im = plt.imshow(a, interpolation='none', aspect='auto', vmin=0, vmax=1)

    def animate_func(i):
        if i % fps == 0:
            print( '.', end ='' )

        im.set_array(snapshots[i])
        return [im]

    anim = animation.FuncAnimation(
                                   fig, 
                                   animate_func, 
                                   frames = nSeconds * fps,
                                   interval = 1000 / fps, # in ms
                                   )

    anim.save(str(ct) + '.gif', fps=fps)

```
